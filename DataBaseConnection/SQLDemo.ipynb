```python
#want to ctreate a table
cr=db.cursor()
create_table_cmd="""CREATE TABLE Persons(PersonID int,
                                 LastName Varchar(255),
                                 FirstName Varchar(255),
                                 Address varchar(255),
                                 City varchar(255))"""
cr.execute(create_table_cmd)
print("table created successfully")
```
