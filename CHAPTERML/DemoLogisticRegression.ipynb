```python
from sklearn.linear_model import LogisticRegression

x=[[1],[2],[3],[4],[5],[6],[7],[8]]
y=[0,0,0,0,1,1,1,1]

model=LogisticRegression()
model.fit(x,y)

print("predict for 4.5 class:",model.predict([[4.6]]))
print("predict for 8 class:",model.predict([[0]]))
print("predict for 9 class:",model.predict([[9]]))
print("predict for 6 class:",model.predict([[-6]]))
```

```python
#annual test third part : SKLEARN
import numpy as np

x=[[1],[2],[3],[4],[5],[6],[7],[8]]
y=[0,0,0,0,1,1,1,1]

#SIGmoid function
def sigmoid(z):
    return 1 /(1 +np.exp(-z))

#logistic regression function
def predict(x,weight=1.5,bias=-6.7):
    z=weight*x+bias

    prob=sigmoid(z)
    return 1 if prob>=0.5 else 0
```

```python
predict(4.6,1.5,-6)
```
