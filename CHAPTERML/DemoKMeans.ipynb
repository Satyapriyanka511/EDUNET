```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('student_clustering.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
#plotting the data points
plt.scatter(df['cgpa'],df['iq'])
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.show()
```

```python
WCSS=[]
for i in range (1,11):
    km=KMeans(n_clusters=i)
    km.fit(df)
    WCSS.append(km.inertia_)

WCSS
```

```python
# Plotting the Elbow Curve
plt.plot(range(1, 11), WCSS, marker= 'o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
x=df.iloc[:, :].values #converting dataframe to numpy array
km=KMeans(n_clusters=4)
y_km=km.fit_predict(x)
```

```python
print(y_km)
```

```python
print(y_km==1)
```

```python
#print(y_km==0)
x[y_km==0,0]  #cgpa of cluster
x[y_km==1,1]  #iq of the cluster 1
```

```python
#displaying the cluster all clusters
plt.scatter(x[y_km==0,0],x[y_km==0,1])
plt.scatter(x[y_km==1,0],x[y_km==1,1])
plt.scatter(x[y_km==2,0],x[y_km==2,1])
plt.scatter(x[y_km==3,0],x[y_km==3,1])
```

```python
#print silhoutte score
from sklearn.metrics import silhouette_score
score=silhouette_score(x,y_km)
print("silhoutee score:",score)
```
