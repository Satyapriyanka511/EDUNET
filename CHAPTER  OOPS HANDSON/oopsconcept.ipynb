```python
#multiple inheritance
#multiple parent class
#one child class

class parent_one:
    saving=250000
    def display(self):
        print("method of parent one")
```

```python
class parent_two:
    property=50000

    def show(self):
        print("method of parent two")
```

```python
class child(parent_one, parent_two):
    def __init__(self):
        print("in a child")
```

```python
ch1=child()
```

```python
ch1=child()
print(ch1.saving) #parent one attribute
print(ch1.property) #parent two attribute
ch1.display()
ch1.show()
```

```python
#encapulation
#wraping a data (variable) in a method

class employee:
    name="priya"
    age=19
    city="kothapeta"

    def __init__(self):
        print("----employee details----")

    def get_name(self):
        return self.name
        
    def get_age(self):
        return self.age
    
    def get_city(self):
        return self.city
    
    #setter
    def set_name(self, new_name):
        self.name=new_name

    def set_age(self, new_age):  
        self.age=new_age

    def set_city(self, new_city):
        self.city=new_city    
```

```python
emp=employee()
#print(emp.name)
print(emp.get_name())
print(emp.get_age())
print(emp.get_city())
print("-----set employee-----")
emp.set_name("satya")
emp.set_age(22)
emp.set_city("kakinada")
print(emp.get_name())
print(emp.get_age())
print(emp.get_city())
```
