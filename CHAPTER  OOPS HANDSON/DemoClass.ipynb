```python
class employee:
    def __init__(self,new_name,new_age,new_city,):
        self.name=new_name
        self.age=new_age
        self.city=new_city
        print("-----new employee-----")
        print("name:",self.name)
        print("age:",self.age)
        print("city:",self.city)
```

```python
emp=employee("priya",19,"amp")
emp=employee("darshini",20,"vizag")
```
