```python
class vehicle:
    def __init__(self,make,model,year):
        self.make=make
        self.model=model
        self.year=year

    def vehicle_info(self):    
        print("-----vehicle details-----")
        print("make:",self.make)
        print("model:",self.model)
        print("year:",self.year)    

    def update_model(self,model):
         self.model=model
```

```python
obj_veh=vehicle("tata","nexon",2020)
print("-----Before Update-----")
obj_veh.vehicle_info()
print("-----After Update-----")
obj_veh.update_model("safari")
obj_veh.vehicle_info()
```
